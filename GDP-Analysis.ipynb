```python
#Analyzing Arab world
df_pr = df[df['Country Name'] == 'Arab World']
df_pr.plot(kind ='line', x = 'Year', y = 'Value', figsize=(10,5))
```

```python
df_pr = df[df['Country Name'] == 'India']
df_pr.plot(kind ='line', x = 'Year', y = 'Value', figsize=(10,5), grid = True)
```

```python
data = df_pr.values

gdp_change = [0]

for i in range(1,len(data)):
    
    prev = data[i-1][3]
    cur = data[i][3]
    
    gdp_change.append(round(((cur - prev) / prev)*100,2))
```

```python
df_pr.assign(GDP = gdp_change)
```
